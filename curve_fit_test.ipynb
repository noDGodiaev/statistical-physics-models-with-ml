```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x, k, b):
    return k*x+b
```

```python
x = np.linspace(0, 4, 5)
```

```python
xx = [1, 2, 3, 3, 4]
```

```python
popt, pcov = curve_fit(func, x, xx, absolute_sigma=True)
popt
```

```python
perr = np.sqrt(np.diag(pcov))
perr
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x, xx)
plt.scatter(x, func(x, *popt))
y1 = (popt[0]-perr[0])*x+popt[1]-perr[1]
y2 = (popt[0]+perr[0])*x+popt[1]+perr[1]
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.fill_between(x, y1, y2, facecolor="gray", alpha=0.25)
plt.grid()
```

```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt
x_data = np.arange(-5, 5, 0.01)
y_data = stats.norm.pdf(x_data, 0, 1)
plt.plot(x_data, y_data)
```

```python
def Gauss(x, A, B):
    y = A*np.exp(-1*B*x**2)
    return y
```

```python
popt, pcov = curve_fit(Gauss, x_data, y_data, absolute_sigma=True)
perr = np.sqrt(np.diag(pcov))
perr
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x_data, y_data)
plt.scatter(x_data, Gauss(x_data, *popt))
y1 = (popt[0]-perr[0])*np.exp(-1 * x_data**2*popt[1]-perr[1])
y2 = (popt[0]+perr[0])*np.exp(-1 * x_data**2*popt[1]+perr[1])
plt.scatter(x_data, y1)
plt.scatter(x_data, y2)
plt.fill_between(x_data, y1, y2, facecolor="gray", alpha=0.25)
plt.grid()
```
